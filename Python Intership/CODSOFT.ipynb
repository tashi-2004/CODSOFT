```python
# Tashfeen Abbasi
# i22-2041
# NUCES,FAST
import random
print("\n")
print("\t\t\t\t\t_____________________________")
print("\t\t\t\t\t|  Rock-Paper-Scissors Game |")
print("\t\t\t\t\t|___________________________|")

me = 0
pc = 0
while True:
    print("\n")
    print("-------------------------------------------------------------------------------------------------------")
    print("_____________________")
    print("|  Select from List |")
    print("|    1. Rock        |")
    print("|    2. Paper       |")
    print("|    3. Scissors    |")
    print("|    4. Exit        |")
    print("|___________________|")
    print("\n")

    mine = input("Enter: ")
    print("-------------------------------------------------------------------------------------------------------")
    if mine.isdigit():
        if 1 <= int(mine) <= 3:
            user_choice_index = int(mine) - 1
            list1 = ['rock', 'paper', 'scissors']
            me_choices = list1[user_choice_index]

            pc_choices = random.choice(['rock', 'paper', 'scissors'])

            if me_choices == pc_choices:

                final = "It is a tie :)"

            elif (me_choices == 'rock' and pc_choices == 'scissors') or \
                 (me_choices == 'scissors' and pc_choices == 'paper') or \
                 (me_choices == 'paper' and pc_choices == 'rock'):
                final = "Me win ! "
                me += 10

            else:
                print("\n")
                final = "Me lost :( "
                pc += 10

            print("My choice  : {}".format(me_choices))
            print("\n")
            print("Pc's choice: {}".format(pc_choices))
            print("\n")
            print("-------------------------------------------------------------------------------------------------------")
            print("\n")
            print(final)
            print("\n")
            print("-------------------------------------------------------------------------------------------------------")
            print("\t\t\t\t\t___________")
            print("\t\t\t\t\t|  Scores |")
            print("\t\t\t\t\t|_________|")
            print("Mine: {}".format(me))
            print("\n")
            print("PC  : {}".format(pc))
            print("\n")
            print("-------------------------------------------------------------------------------------------------------")
            tashi_again = input("Do you want to play again?").lower()
            if tashi_again != 'yes' and tashi_again != 'y':
                print("\n")
                print("Bye Bye :) !")
                break
        elif mine == '4':
            print("\n")
            print("Bye Bye!")
            break
        else:
            print("Please enter 1, 2, 3, or 4.")
    else:
        print("Please enter a numerical value.")
```

```python
# Tashfeen Abbasi
# i22-2041
# NUCES,FAST
tashi_list = []
while True:
    print("\n")
    print("\t\t\t\t\t_______________________________________")
    print("\t\t\t\t\t|  Tashi's Contact Management System  |")
    print("\t\t\t\t\t|_____________________________________|")
    print("_______________________")
    print("|  1. Add Contact     |")
    print("|  2. View Contacts   |")
    print("|  3. Search Contact  |")
    print("|  4. Update Contact  |")
    print("|  5. Delete Contact  |")
    print("|  6. Exit            |")
    print("|_____________________|")
    print("\n")
    tashi_choice = input("Enter your choice (1/2/3/4/5/6): ")
    print("\n")
    if tashi_choice == '1':
        name = input("Enter contact name: ")
        print("\n")
        phone = input("Enter phone number: ")
        print("\n")
        email = input("Enter email address: ")
        print("\n")
        address = input("Enter address: ")
        print("\n")
        new_contact = {'name': name, 'phone': phone, 'email': email, 'address': address}
        tashi_list.append(new_contact)
        print("\t\t\t\t\tContact '{}' added successfully!".format(name))

    elif tashi_choice == '2':
        if not tashi_list:
            print("No contacts found.")
        else:
            print("\t\t\t\t\t\t__________________")
            print("\t\t\t\t\t\t|  Contact List  |")
            print("\t\t\t\t\t\t|________________|")
            print("\n")
            for i in tashi_list:
                print("Name: {}, Phone No: {}".format(i['name'], i['phone']))
                print("\n")

    elif tashi_choice == '3':
        tashi_search = input("Enter name or phone number to search: ")
        print("\n")
        matching_contacts = [
            contact for contact in tashi_list
            if "{}".format(tashi_search.lower()) in "{}".format(contact['name'].lower()) or tashi_search in contact['phone']
        ]

        if not matching_contacts:
            print("No matching contacts found.")
        else:
            print("\t\t\t\t\t\t_______________________")
            print("\t\t\t\t\t\t|  Matching Contacts  |")
            print("\t\t\t\t\t\t|_____________________|")
            print("\n")
            for i in matching_contacts:
                print("Name: {}, Phone: {}".format(i['name'], i['phone']))

    elif tashi_choice == '4':
        name = input("Enter the name of the contact to update: ")
        print("\n")
        for i in tashi_list:
            if i['name'].lower() == name.lower():
                i['phone'] = input("Enter the new phone number: ")
                print("\n")
                i['email'] = input("Enter the new email address: ")
                print("\n")
                i['address'] = input("Enter the new address: ")
                print("\n")
                print("\t\t\t\t\tContact '{}' updated successfully!".format(name))
                break
        else:
            print("Contact '{}' not found.".format(name))

    elif tashi_choice == '5':
        name = input("Enter the name of the contact to delete: ")
        print("\n")
        for i in tashi_list:
            if i['name'].lower() == name.lower():
                tashi_list.remove(i)
                print("\t\t\t\t\tContact '{}' deleted successfully!".format(name))
                break
        else:
            print("Contact '{}' not found.".format(name))

    elif tashi_choice == '6':
        print("\t\t\t\t\tExiting Tashi's Contact Management System!")
        break

    else:
        print("\t\t\t\t\tWrong choice! ")
        print("\n")
```

```python
# Tashfeen Abbasi
# i22-2041
# NUCES,FAST
import random
import string
print("\t\t\t\t\t\t________________________________")
print("\t\t\t\t\t\t|  Tashi's Password Generator  |")
print("\t\t\t\t\t\t|______________________________|")
print("\n")
while True:
    tashi_length = input("Enter the length of the password: ")
    if tashi_length.isdigit() and int(tashi_length) > 0:
        tashi_length = int(tashi_length)
        break
    else:
        print("Please enter a valid positive integer.")


tashi_ch = string.ascii_letters + string.digits + string.punctuation
tashi_pwd = ""
for i in range(tashi_length):
    tashi_pwd += random.choice(tashi_ch)

print("\n\n")
print("\t\t\t\t\t\t__________________________")
print("\t\t\t\t\t\t|   Generated Password   |")
print("\t\t\t\t\t\t|________________________|")
print("\n")
tashi_pwd
```
